Using parashatrans to Define Parashat Vayeshev. The parashatrans feature allows direct identification of Parashat Vayeshev without manually specifying chapters.

```python
# Get all nodes for Parashat Vayeshev using the parashatrans feature
parashatNodes = [n for n in F.otype.s('verse') if F.parashanum.v(n) == 9]
```

## Query 1: Searching for Root "נכר" (Recognize)

This query uses the parashatrans feature to limit results to Parashat Vayeshev.

```python
# Find all words with the root 'נכר' in Parashat Vayeshev
results_kafresh = []

for word in F.otype.s('word'):
    if F.lex.v(word) == 'NKR[':
        verse = L.u(word, 'verse')
        if verse[0] in parashatNodes:  # Check if the word belongs to Parashat Vayeshev
            clause = L.u(word, 'clause')  # Get the containing clause
            results_kafresh.append({
                "location": T.sectionFromNode(word),
                "clause": T.text(clause, fmt="text-orig-full")
            })

# Output results
for result in results_kafresh:
    loc = result["clause"]
    print(f"Found 'כר' in {loc[0]} {loc[1]}:{loc[2]} - clause: {result['clause']}")
```
